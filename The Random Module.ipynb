```python
# Import the random module
import random
```

```python
# Use .randint to generate an interger inbetween the specified range
random.randint(-90,90)
```

```python
# to obtain a random latitute we can include the .random function which includes decimal floats from 0 to 1
# we need to change the .randint to 89 since the max is 90
random_number = random.randint(-90,89) + random.random()
random_number
```

```python
# We can create a empty list and loop which includes our function above to populate the list with x # of numbers 
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
```

```python
# .randrange generates a random interger from the range, based on the "step" increment
# can be thought of as generating an integer where the random # must be in the range and divisible by the step
random.randrange(-90, 90, step=5) 
```

```python
# random.uniform generates an random floating point decimal within the range
random.uniform (-90,90)
```

```python
# We can also create a function asks to specify a size then runs a loop to populate an array with random numbers 
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500 and use the timit function to see how long it takes (this takes milliseconds)
%timeit latitudes(1500)
```

```python
# Import the NumPy module so we can use NumPy's built in random module rather than importing the random module above 
import numpy as np

# Import timeit.
import timeit

# the .random.uniform function, allows us to create an array of size = x, of random values with the lower and uper limit 
# timeit shows this takes microseconds and is quicker 
%timeit np.random.uniform(low=-90, high=90,size=1500)

```
